<div dir=rtl>
    در مثال فوق، ما دو پارامتر را برای تابع در نظر گرفتیم که به یکی از آن ها مقدار پیش فرض اولیه دادیم. به این معنا که اگر کاربر مقداری را برای آن وارد نکرد، آن مقداری را که ما برایش در نظر گرفتیم، در تابع اعمال شود.
    اما باید در نظر داشت که اگر پارامتر یا پارامترهایی باید مقدار دهی اولیه شوند، باید آن ها را در آخر آرگومان ها نوشت
</div>

```python
def person(lname='Alavi',fname):
    print(f"hello. I'm {fname} {lname}.")

person('Ali')
```


<div dir=rtl>
    همانطور که در مثال فوق مشاهده می کنید، از آن جا که آرگومانی را که مقدار اولیه دارد در ابتدا نوشتیم و در فراخوانی تابع فقط یک مقدار را وارد کردیم(که برای همان آرگومانی بود که مقدار اولیه نداشت) اما برنامه دچار خطا شد. پس برای همین است که آرگومان هایی که مقدار اولیه ندارند در ابتدا و سپس آن هایی که مقدار اولیه دارند نوشته می شود.
    هم چنین ما می توانیم با استفاده از نام آرگومان ها و بدون هیچ ترتیبی، یک تابع را فراخوانی کرده و آرگومان های آن را مقدار دهی کنیم:
    
</div>

```python
def person(fname,city,year=1980,lname='Alavi'):
    import datetime
    now=datetime.datetime.now()
    s1=f"Hello. I'm {fname} {lname}."
    s2=f"I'm from {city}."
    age=now.year-year
    s3=f"I'm {age} years old."
    print(s1+"\n"+s2+"\n"+s3+"\n")

person('Ali','Shiraz',1997)
person(fname='Reza',
       year=1994,
       lname='Darbandi',
       city='Tehran')
person(city='Aliabad',fname='Naqi',lname='Mamoli')
```

<div dir=rtl>
    همچنین ما می توانیم یک tuple را به عنوان آرگومان های ورودی به یک تابع پاس بدهیم:
</div>

```python
def person(*args):
    s1, s2, s3 ="","","",
    if len(args)>=2:
        s1=f"hello. I'm {args[0]} {args[1]}."
    if len(args)>2:
        s2=f"I'm from {args[2]}"
    if len(args)>3:
        age=2021-args[3]
        s3=f"I'm {age} years old."
    print(s1+"\n"+s2+"\n"+s3+"\n")

person('Ali','Rezaei','Shiraz',1989)
person('Hassan','Hassani')
person('arasto','amel','aliabad')
person()
```

<div dir=rtl>
    در مثال فوق نام tuple را می توانستیم مقدار دلخواه دیگری به جز args قرار دهیم.
    به این روش ما می توانیم به تعداد دلخواهی آرگومان ورودی به تابع ارسال کنیم.
    همچنین با کمک متد len می توانیم به تعداد آرگومان های ورودی دسترسی داشته باشیم و بر اساس آن نیز شرایطی را مانند مثال فوق اِعمال کنیم.
</div>
<div dir=rtl>
    همچنین می توانیم یک دیکشنری را به عنوان آرگومان ورودی به تابع ارسال کنیم:
</div>

```python
def person(**dic):

    s1=f"hello. I'm {dic['fname']} {dic['lname']}." if 'fname' in dic else ''
    s2=f"I'm from {dic['city']}" if 'city' in dic else '' 
    s3=f"I'm {2021-dic['year']} is old" if 'year' in dic else ''
    print(s1+"\n"+s2+"\n"+s3+"\n")

person(fname='Karim',
       lname='Karimi',
       city='Kerman',
       year=1974
       )
person(lname='mamoli',
       fname='naqi')
```

<div dir=rtl>
    در مثال بالا می بینید که با نوشتن دو ستاره توانستیم اطلاعات را به صورت دیکشنری دریافت کرده و با استفاده از key، از آن ها در تابع استفاده کردیم. اما باید این نکته را در نظر داشته باشید که برای فراخوانی تابع در بدنه اصلی برنامه، باید با نام همان key ها، آرگومان را مقدار دهی کرد.
</div>
<div dir=rtl>
آیا تابع می تواند پس از اجرا، مقدار یک متغیر را تغییر دهد؟
</div>

```python
a=3
b=4
x=5
y=6
def swap_ab(a,b):
    a,b=b,a
    print(a,b)
ls=[x,y]
def swap_xy(ls):
    ls[0],ls[1]=ls[1],ls[0]
    print(ls)
print(a,b)
swap_ab(a,b)
print(a,b)
print("------")
print(ls)
swap_xy(ls)
print(ls)
```

<div dir=rtl>
    در مثال فوق می بینیم که تابع اولی در درون خود مقدار دو متغیر را باهم جابه جا و چاپ می کند. اما در بیرون از تابع و بعد از اجرای تابع اول، مقادیر آن دو متغیر مجدد چاپ می شوند که مشخص می کند مقدارشان باهم جابه جا نشده است.
    اما در تابع دوم، ما متغیرهایی که میخواستیم مقادیرشان باهم جابه جا شود را در یک لیست قرار دادیم و سپس آن را به تابع ارسال کردیم. بعد از اجرای تابع دوم نیز مجددا مقدار لیست را چاپ کردیم که مشخص کرد این تغییر اعمال شده است.
</div>
